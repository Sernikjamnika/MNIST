# Softmax
## Making a model

tf.placeholder returns tensor of shape (None, 784) (so it's a vector) with values of type float32
<ul>
    <li> $x$ given instance </li>
    <li> $W$ vector of weights - tensor of shape 784 x 10</li>
    <li> $b$ bias - vector of 10 zeros </li>
</ul>

```python
x = tf.placeholder(tf.float32, [None, 784])
W = tf.Variable(tf.zeros([784,10]))
b = tf.Variable(tf.zeros([10]))
```

Let's implement softmax model. A generalized Logistic Regression. thanks to it we don't have to train multiple binary classifiers. It's also called Multinomial Logistic Regression.<br>
How does it work?<br>
It computes the score for each class then estimates the probability of each class by applying the softmax function - normalized exponential. After that we will use $argmax()$ to search for our prediction - we are taking class with highest probability.<br>
Softmax score for class k:
\begin{equation}
    s_{k}(x) = \theta_{k}^{T} * x
\end{equation}
<ul>Where:
    <li> $\theta_{k}^{T}$ is a dedicated parameter vector for each class transposed.</li>
    <li> $x$ is a give instance in our case a vector of the photo. </li>
  </ul> 
Softmax function: <br>
\begin{equation} 
    \hat{p}_{k} = \frac{exp(s_k(x))}{\sum_{j=1}^{K} \exp(s_j(x))}
\end{equation}
<ul>Where:
    <li> K is the number of classes in our case 10.</li>
    <li> $s(x)$ is a vector containing the scores for each class (what our model predicted).</li>
    <li> $\hat{p}_{k}$ is the estimated probability that the instance $x$ belongs to the class $k$ - scores for each class for an instance.</li>
<\ul>


```python
y = tf.nn.softmax(tf.matmul(x, W) + b)
```

softmax returns tensor of the same shape as tf.matmul(x, W) + b<br>
tf.matmul multiplies tensors 

## Training

To determine loss of our model we will use <b>cross - entropy</b>.<br>
This function based on information theory. It measures distance between predicted value and the real one output. It's often used with problems such as MNIST where model is multilabel but not multioutput. So it means it will be our <b>loss function</b>.<br>
It's given by equation: <br>
\begin{equation*}
J(\Theta) = -\frac{1}{m}\sum_{i=1}^{m}\sum_{k=1}^{K}y_k^{(i)}\log(\hat{p}_{k})
\end{equation*}
Where $y$ is our predicted probability distribution, and $y'$ is the true distribution (the one-hot vector with the digit labels).

```python
y_ = tf.placeholder(tf.float32, [None, 10])
cross_entropy = tf.reduce_mean(-tf.reduce_sum(y_ * tf.log(y), axis=[1]))
```

```python
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy)
```

train_step uses Gradient Descent with learning rate 0.1 to minimize cross_entropy <br>


In for loop we divide our train set into batches randomly chosen of 100 examples in each.<br>
Batch_xs is a batch with images and batch_ys are labels<br>
Then Interactive Session feeds train_step, so Gradient Descent algorithm, with batches. This kind of feeding our model with small bunches of data is called stochastic training.

```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
for _ in range(10000):
  batch_xs, batch_ys = mnist.train.next_batch(100)
  sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```

## Evaluating model

```python
correct_prediction = tf.equal(tf.argmax(y,1), tf.argmax(y_,1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
print(sess.run(accuracy, feed_dict={x: mnist.test.images, y_: mnist.test.labels}))
prediction = tf.argmax(y,1)
labels = tf.argmax(y_, 1)
predicted_results = prediction.eval(feed_dict={x: mnist.test.images}, session=sess)
real_results = labels.eval(feed_dict={y_: mnist.test.labels}, session=sess)
print("ImageId, Label")
for i in range(len(predicted_results)):
    print("{0} {1} {2}". format(i, predicted_results[i], real_results[i]))
```

As a conclusion we can see that without fine tuning the result is pretty bad comparing to what Decision Forest or even Convultional Neural Network can do. Despite this fact the great advantage of this approach is the fact that it is much more understandable than neural network.
